```python
import numpy as np
import matplotlib . pyplot as plt
import pandas as pd
from sklearn . model_selection import train_test_split
from sklearn . linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_excel (r"C:\Users\17b-132-se\Desktop\Salary.xlsx")
x = dataset ["Experience"]. values . reshape ( -1 ,1)
y = dataset ["Salary"]. values . reshape ( -1 ,1)
```

```python
plt.plot(dataset["Experience"], dataset["Salary"], 'o')
plt.title("Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
xTrain , xTest , yTrain , yTest = train_test_split (x ,y , test_size =0.2 ,random_state = 0)
```

```python
linearRegressor = LinearRegression ()
linearRegressor . fit ( xTrain , yTrain )
```

```python
yPrediction = linearRegressor . predict ( xTest )
```

```python
yPrediction = pd.DataFrame ({'Actual': yTest . flatten () , 'Predict': yPrediction . flatten () })
yPrediction 
```

```python
import numpy as np

xTrain = np.array(xTrain).reshape(-1, 1)
yTrain = np.array(yTrain).reshape(-1, 1)

plt.scatter(xTrain, yTrain, color='red')
plt.plot(xTrain, linearRegressor.predict(xTrain), color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
import numpy as np

xTrain = np.array(xTest).reshape(-1, 1)
yTrain = np.array(yTest).reshape(-1, 1)

plt.scatter(xTest, yTest, color='red')
plt.plot(xTrain, linearRegressor.predict(xTest), color='blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
